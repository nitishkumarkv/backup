```python
num_train = positron_features_data_2.beamEnergy.size
num_train
```

```python
X_pre = np.stack(positron_features_data_2.head(num_train).lay_data.values)
print(X_pre.shape)
X_pre
```

```python
X_pre[:,26]+X_pre[:,27]
```

```python
#X = X_pre
X = np.c_[np.ones(num_train), X_pre[:,:24], X_pre[:,24]+X_pre[:,25]+X_pre[:,26]+X_pre[:,27]]
#X = X_pre[:,:]
#X = np.flip(X, axis=1)
X.shape
```

```python
Y = (positron_features_data_2.head(num_train).trueBeamEnergy.values).reshape(num_train,1)
print(Y)
Y.shape
```

```python
theta = np.matmul( np.linalg.pinv( np.matmul(X.T, X) ) , np.matmul(X.T, Y) )
print(theta.shape)
theta
```

```python
MSE = np.matmul((Y-np.matmul(X, theta)).T, (Y-np.matmul(X, theta)))/(num_train-X.shape[1])
MSE[0,0]
```

```python
sigma_squared_theta =  MSE[0,0] * np.linalg.pinv( np.matmul(X.T, X) )
sigma_squared_theta
```

```python
err_theta = np.sqrt(np.array([sigma_squared_theta[i,i] for i in range(X.shape[1])]))
err_theta
```

```python
lay = [i for i in range(1,X.shape[1])]

plt.figure(figsize=[10,5])

#plt.plot([28.5,28.5], [-0.75, 1],  linestyle="-", color="r", markersize=10)
#plt.plot([40.5,40.5], [-0.75, 1],  linestyle="-", color="r", markersize=10)

plt.errorbar(lay, theta[1:], yerr = err_theta[1:], marker = ".", linestyle="--", color="b", markersize=7)


plt.xlabel("Layer Number")
plt.ylabel("Weight")
plt.title("Layerwise weights for Positron Energy Reconstruction")


#plt.text(12, 0.5, 'ECAL', fontsize=20, horizontalalignment='center',verticalalignment='center')
#plt.text(34.5, 0.5, 'HCAL', fontsize=20, horizontalalignment='center',verticalalignment='center')
#plt.text(62, 0.5, 'AHCAL', fontsize=20, horizontalalignment='center',verticalalignment='center')


plt.grid(linestyle='--')
plt.legend()
plt.show()
```
