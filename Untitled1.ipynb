```python
pion_features_data = pd.read_pickle("new_combined_data_pion_ene_std.pkl")
pion_features_data
```

```python
def features(df):
    return np.array(list(df.lay_energy) + list(df.totEne_std)  )
```

```python
pion_features_data["features"] = pion_features_data.apply(features, axis=1)
pion_features_data
```

```python
def squ_by_true(y_true, y_pred):
    error = y_true - y_pred
    squared_loss_by_true = tf.square(error) / y_true
    return squared_loss_by_true
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import scipy.stats as scs
from scipy.optimize import curve_fit
```

```python
def gaussian(x,  mean,a, sigma):
    return a * np.exp(-((x - mean)**2 / (2 * sigma**2)))

pred = x_test_set.model1_pred.values
true = x_test_set.trueBeamEnergy.values

fracarr = (pred)/true

bin_heights, bin_borders, _ = plt.hist(fracarr,range=[.5,+1.5], bins=100, label='histogram')

bin_centers = bin_borders[:-1] + np.diff(bin_borders) / 2
popt, pcov = curve_fit(gaussian, bin_centers, bin_heights, p0=[1., 1000., 0.1],bounds = ([-np.inf,0,0],[np.inf,np.inf,np.inf]))
perr = np.sqrt(np.diag(pcov))
print(popt,perr)

x_interval_for_fit = np.linspace(bin_borders[0], bin_borders[-1], 100)
plt.plot(x_interval_for_fit, gaussian(x_interval_for_fit, *popt), label='fit')
plt.legend()


plt.xlabel('pred - true / true')
plt.ylabel('counts')
#plt.title(r'$\mathrm{pred - true / true:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
plt.title(r'$\mathrm{pred - true / true:}\ \mu=%.3f,\ \sigma=%.3f,\ E=$%i to %i' %(popt[0], popt[2],0,300))
plt.grid(True)
    
plt.show()
#    return [popt[0], popt[2],perr[0], perr[2]]
```
