```python
x = np.array([[1000,1000,0.01], [2000,1000,20], [1000,3000,0.01]])
y = np.array([[300], [400], [500]])
print(x)
print(y)
print(np.matmul(x.T,y))
```

```python
xt_x = np.matmul(x.T,x)
xt_x_inv = np.linalg.pinv(xt_x)
print(xt_x)
print(xt_x_inv)
print(np.matmul(xt_x_inv, xt_x))
```

```python
print(np.matmul(xt_x_inv, np.matmul(x.T,y)))
```
